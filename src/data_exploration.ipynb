## 1.3 Data Types and Missing Values

```python
#Type of data in the Dataset
data.dtypes
```

```python
#Number of missing values in each column
missing_values = data.isnull().sum()

#Percent of the values of each columns over the total values
missing_percent = (missing_values /len(data))*100

#Presenting missing data as a dataframe
missing_data = pd.DataFrame({"Total values missing": missing_values, "Percentage": missing_percent})

#Sorting missing data from more missing values to less (descending)
missing_data = missing_data[missing_data["Total values missing"] > 0].sort_values(by = "Total values missing", ascending = False)

missing_data
```

```python
#YEAR

nat_count_year = data['Year'].isna().sum()
num_uniq_year = data['Year'].nunique()
print(nat_count_year, num_uniq_year)
```

## 1.4 Univariate analysis

```python
sns. set_style("whitegrid")

#Univariatr analysis of 'Type'column

plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Type', order = data['Type'].value_counts().index)
plt.title('Distribution of Shark Attack Types')
plt.xlabel('Count')
plt.ylabel('Type')
plt.show()
```
