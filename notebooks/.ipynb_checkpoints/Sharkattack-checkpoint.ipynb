```python
def inspect_missing_values(data):

    missing_values = data.isnull().sum()
    missing_percent = (missing_values /len(data))*100
    missing_data = pd.DataFrame({"Total values missing": missing_values, "Percentage": missing_percent})
    missing_data = missing_data[missing_data["Total values missing"] > 0].sort_values(by = "Total values missing", ascending = False)
    
    return missing_data

missing_data = inspect_missing_values(data)
missing_data
```

```python
def clean_data(data):
    
    #dropping columns with 99% missing values
    cleaned_data = data.drop(columns = ['Unnamed: 22', 'Unnamed: 23'])
    
    #Reeplacing cells with "xx" values in 'Case Number' column with NaN
    cleaned_data['Case Number'] = cleaned_data['Case Number'].replace(["xx", '0'],np.nan)
    
    # TODO: Maira
    cleaned_data = cleaned_data.dropna(how="all")
    
    return cleaned_data

cleaned_data = clean_data(data)
cleaned_data
```
